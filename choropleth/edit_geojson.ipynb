```python
import pandas as pd
import numpy as np
import json

data_owid = "data/owid-covid-data.csv"

data_owid_df = pd.read_csv(data_owid)

data_owid_df.rename(columns= {"location": "country_name"}, inplace= True)
```

```python
data_owid_df.head()
```

```python
data_owid_df.info()
```

```python
data_owid_df['date']= pd.to_datetime(data_owid_df['date'])

max_date = data_owid_df.groupby('country_name')['date'].max()
max_date.rename("max_date", inplace= True)
max_date.head()
```

```python
df = pd.merge(data_owid_df, max_date, on='country_name', how='outer')
df[df.iso_code == 'USA']
```

```python
recent = df[df.date == df.max_date]
recent = recent[['iso_code', 'country_name', 'continent', 'total_cases_per_million', 'total_deaths_per_million', 'total_vaccinations_per_hundred', 'gdp_per_capita']]
recent.head()
```

```python
recent_dict = recent.to_dict(orient='records')
recent_dict
```

```python
with open('countries.geojson','r') as f:
    data = json.loads(f.read())
features = data['features']
```

```python
geo = pd.DataFrame(features)
geo.head()
```

```python
pd.DataFrame(geo['properties'][1], index = [0])
```

```python
list = []

for x in range(0,len(geo['properties'])):
    country_data = pd.DataFrame(geo['properties'][x], index = [x])
    list.append(country_data)
properties = pd.concat(list).rename(columns = {'ADMIN': 'country_name', 'ISO_A3' : 'iso_code'})
properties = properties['iso_code']
properties.head()
```

```python
merged = pd.merge(properties, recent, on = ['iso_code'], how = 'left')
merged.head()
```

```python
merged[merged.iso_code == 'USA']
```

```python
properties_new = merged.to_dict(orient='records')
properties_new
```

```python
geo['properties_new'] = properties_new
geo.head()
```

```python
geo_new = geo[['type', 'properties_new', 'geometry']].rename(columns = {'properties_new': 'properties'})
geo_new.head()
```

```python
features_new = geo_new.to_dict(orient='records')
data['features'] = features_new
# data
```

```python
with open('country_properties.geojson', 'w') as f:
    json.dump(data, f)
```
