```python
import plotly
import plotly.graph_objects as go

import pandas as pd

# Load data
df = pd.read_csv("../data/owid-covid-data.csv")
# df.columns = [col.replace("AAPL.", "") for col in df.columns]
df.head()
```

```python
country_df = df[df["location"] == "United States"]
# Create figure
fig = go.Figure()

fig.add_trace(
    go.Scatter(x=list(country_df.date), y=list(country_df.total_vaccinations_per_hundred), yaxis="y1", name="Vaccinations"))


fig.add_trace(
    go.Scatter(x=list(country_df.date), y=list(country_df.total_cases_per_million), yaxis="y2", name="Cases"))

```

```python

# Set title
fig.update_layout(
    title_text="Time series with range slider and selectors"
)

# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    ),
    yaxis=dict(
        anchor="x",
        autorange=True,
        domain=[0, 0.5],
        linecolor="#673ab7",
        mirror=True,
        # range=[df_country.total_vaccinations_per_hundred.min(), df_country.total_vaccinations_per_hundred.max()],
        showline=True,
        side="right",
        tickfont={"color": "#673ab7"},
        tickmode="auto",
        ticks="",
        titlefont={"color": "#673ab7"},
        type="linear",
        zeroline=False
    ),
    yaxis2=dict(
        anchor="x",
        autorange=True,
        domain=[0.5, 1],
        linecolor="#E91E63",
        mirror=True,
        # range=[df_country.total_cases_per_million.min(), df_country.total_cases_per_million.max()],
        showline=True,
        side="right",
        tickfont={"color": "#E91E63"},
        tickmode="auto",
        ticks="",
        titlefont={"color": "#E91E63"},
        type="linear",
        zeroline=False
    )
)

fig.show()
```

```python
plotly.offline.plot(fig,filename='cumulative.html',config={'displayModeBar': True})
```
